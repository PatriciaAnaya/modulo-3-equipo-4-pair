```python
import pandas as pd
```

```python
# Paso 1: Abrir el archivo y mostrar el dataframe

# Este conjunto de datos proporciona información diversa sobre estos productos.
df_rating = pd.read_csv('datasets/rating-and-performance.csv', index_col=0
                        )
df_rating
```

```python
df_rating.head()
```


Cread una nueva columna donde tengamos la diferencia entre el precio al que vende el e-commerce menos el precio del minorista. Insertad esta columna en la posición dos.

```python
# Insertar la columna en la posición dos
df_rating.insert(2, 'diferencia_precios', (df_rating['price'] - df_rating['retail_price']))
```

```python
df_rating.head()
```


Usando el loc:

Extraed todas las filas de la columna "price". ¿De qué tipo es el resultado obtenido?

```python
print(df_rating.loc[:,"price"])

```


Extraed todas las filas de las columnas "price" y "retail_price". ¿De qué tipo es el resultado obtenido?

```python
price_loc = df_rating.loc[:,["price", "retail_price"]]
```

```python
price_loc.info()
```

Extraed las filas pares, de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count".

```python
filas_pares_loc = df_rating.loc[::2,["rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count"]]
filas_pares_loc.head()
```

```python
filas_impares_loc = df_rating.loc[1::2,["rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count"]]
filas_impares_loc.head()
```


Extraed las filas cuyo precio sea mayor que 10 y sus unidades vendidas ("units_sold") sea mayor que 5000.

```python
filtro_df_loc = df_rating.loc[(df_rating['price'] > 10) & (df_rating['units_sold'] > 5000)]
filtro_df_loc.head()
```


Usando el iloc (sacaremos los mismos resultados que con el loc):

-Extraed todas las filas de la columna "price".

-Extraed todas las filas de las columnas "price" y "retail_price".

-Extraed las filas pares, de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count".

```python
# Usando iloc (los mismos resultados que con loc):

# Extraer todas las filas de la columna "price"
prices_iloc = df_rating.iloc[:, 2]
prices_iloc

# Extraer todas las filas de las columnas "price" y "retail_price"
prices_retail_iloc = df_rating.iloc[:, [2, 3]]
prices_retail_iloc

# Extraer las filas pares de las columnas especificadas
filas_pares_iloc = df_rating.iloc[::2, 7:14]
filas_pares_iloc.head()
```
