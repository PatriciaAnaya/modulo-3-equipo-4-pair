## 1. Resumen básico de indicadores clave

1. Calcula, para cada una de estas variables:

    - Desviación estándar

```python
#population

varianza_population = df['population'].var()
desv_std_population = df['population'].std()
rango_population = df['population'].max() - df['population'].min()
max_population = df['population'].max()
min_population = df['population'].min()
media_population = df['population'].mean()
mediana_population = df['population'].median()
cv_population = desv_std_population / media_population * 100
IQR = df['population'].quantile(0.75) - df['population'].quantile(0.25) # mide la dispersión del 50% central de los datos. Se calcula como la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1).


print(f"La media de los ingresos de los clientes es: {np.round(media_population, 2)}")
print(f"La mediana de los ingresos de los clientes es: {np.round(mediana_population, 2)}")
print(f"La varianza  de los ingresos de los clientes es: {np.round(varianza_population, 2)}")
print(f"El máximo de los ingresos de los clientes es: {max_population}")
print(f"El mínimo de los ingresos de los clientes es: {min_population}")
print(f"La desviación estándar de los ingresos de los clientes es: {np.round(desv_std_population, 2)}")
print(f"El rango de los ingresos de los clientes es: {np.round(rango_population, 2)}")
print(f"Coeficiente de Variación: {cv_population:.2f}%")
print(f"Rango Intercuartílico: {IQR:.2f}")
```
