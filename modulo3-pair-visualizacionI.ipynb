```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv("./datasets/merchant_fill_nonulls.csv") 
```


¿Cuál es la relación entre el precio y la calificación promedio de los productos?

```python
sns.scatterplot(x = "price", 
                y = "mean_product_ratings_count", 
                data = df, 
                size = 10, 
                marker= "v", 
                color = "orange")


# cambiamos los nombres de los ejes
plt.xlabel("Precio")
plt.ylabel("Valoración media producto");
```


¿Cómo se distribuyen los precios de los productos en función de si se utilizan anuncios promocionales (uses_ad_boosts) o no?

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='uses_ad_boosts', y='price', data=df)

plt.title('Distribución de Precios: Con vs. Sin Ad Boosts')
plt.xlabel('¿Usa Ad Boost?')
plt.ylabel('Precio')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='price', hue='uses_ad_boosts', fill=True, common_norm=False, palette='viridis')

plt.title('Densidad de Probabilidad del Precio según Uso de Anuncios')
plt.xlabel('Precio')
plt.ylabel('Densidad')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='uses_ad_boosts', y='price', data=df, split=True, inner="quart")

plt.title('Comparación de Distribución de Precios (Violin Plot)')
plt.show()
```


¿Existe una relación entre el uso de anuncios y las unidades vendidas?

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='uses_ad_boosts', y='total_units_sold', data=df)

plt.title('Unidades vendidas: Con vs. Sin Ad Boosts')
plt.xlabel('¿Usa Ad Boost?')
plt.ylabel('Unidades vendidas')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='total_units_sold', hue='uses_ad_boosts', fill=True, common_norm=False, palette='viridis')

plt.title('Unidades Vendidas según Uso de Anuncios')
plt.xlabel('Unidades Vendidas')
plt.ylabel('Densidad')
plt.show()
```

```python
# comenzamos empezando estableciendo nuestra gráfica y indicando el número de filas y columnas que queremos
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (20,5))

# creamos un boxplot, recordemos que en el caso de Seaborn para indicar en el subplot lo queremos tenemos que usar el parámetro 'ax' dentro del método de la función
# fijaos que la única diferencia con los boxplot que hemos creado previamente es que hemos añadido el parametro 'y' 
sns.boxplot(x = "uses_ad_boosts", 
            y = "total_units_sold", 
            data = df, 
            ax = axes[0], 
            palette= "mako")


# creamos nuestro primer barplot. 📌 LO QUE NOS MUESTRA ES LA MEDIA DE EDAD EN FUNCIÓN DEL ESTADO CIVIL
sns.barplot(x = "uses_ad_boosts", 
            y = "total_units_sold", 
            data = df,
            ci = None,
            ax = axes[1], 
            palette = "magma")


# añadimos título al conjunto de todas las gráficas usando '.suptitle()'
plt.suptitle("Comparación unidades vendidas y uso de publicidad")

# añadimos el 'plt.tigth_layout()' para que se ajusten los elementos de la gráfica
plt.tight_layout()
```

¿Cuál es la distribución de las calificaciones de los empleados (rating_employee) para los productos que tienen el distintivo "Buena calidad del producto" (badge_product_quality) y para los que no lo tienen? Haced que las dos gráficas tengan la misma escala en los dos ejes.

```python
df.groupby('badge_product_quality')['rating_employee'].count()
```

```python
# Filtrar correctamente por categoría
buena = df[df['badge_product_quality'] == 'Buena']['rating_employee']
excelente = df[df['badge_product_quality'] == 'Excelente']['rating_employee']

# Límites comunes
xmin = min(buena.min(), excelente.min())
xmax = max(buena.max(), excelente.max())

bins = 20

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Histograma Buena
axes[0].hist(buena, bins=bins)
axes[0].set_title('Productos con calidad Buena')
axes[0].set_xlabel('Rating del empleado')
axes[0].set_ylabel('Frecuencia')
axes[0].set_xlim(xmin, xmax)

# Histograma Excelente
axes[1].hist(excelente, bins=bins)
axes[1].set_title('Productos con calidad Excelente')
axes[1].set_xlabel('Rating del empleado')
axes[1].set_xlim(xmin, xmax)

plt.tight_layout()
plt.show()

```

```python
buena = df[df['badge_product_quality'] == 'Buena']['rating_employee']
excelente = df[df['badge_product_quality'] == 'Excelente']['rating_employee']

plt.figure(figsize=(8, 5))
plt.boxplot(
    [buena, excelente],
    labels=['Buena', 'Excelente'],
    showfliers=True
)

plt.xlabel('Calidad del producto')
plt.ylabel('Rating del empleado')
plt.title('Distribución del rating del empleado según calidad del producto')
plt.show()
```

Compara diferentes distribuciones de precios de productos en función de si tienen el distintivo "Producto local" o "Buena calidad del producto". Tendréis que crear dos gráficas en un subplot, en la primera poner los precios de los productos locales ("badge_local_product")

```python
# --- Preparación de los datos ---

# Precios según si el producto es local
precio_local = df[df['badge_local_product'] == 'Sí']['price']
precio_no_local = df[df['badge_local_product'] == 'No']['price']

# Precios según calidad del producto
precio_buena = df[df['badge_product_quality'] == 'Buena']['price']
precio_excelente = df[df['badge_product_quality'] == 'Excelente']['price']


# --- Creación de la visualización ---

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Subplot 1: Producto local vs no local
axes[0].boxplot(
    [precio_local, precio_no_local],
    labels=['Local', 'No local'],
    showfliers=True
)
axes[0].set_title('Distribución de precios según producto local')
axes[0].set_xlabel('Tipo de producto')
axes[0].set_ylabel('Precio')

# Subplot 2: Calidad del producto
axes[1].boxplot(
    [precio_buena, precio_excelente],
    labels=['Buena', 'Excelente'],
    showfliers=True
)
axes[1].set_title('Distribución de precios según calidad del producto')
axes[1].set_xlabel('Calidad del producto')

plt.tight_layout()
plt.show()
```
