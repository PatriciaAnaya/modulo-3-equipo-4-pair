```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv("./datasets/merchant_fill_nonulls.csv") 
```

```python
df.head(2)
```

Contesta a las siguientes preguntas utilizando la visualización más apropiada.

¿Cómo se distribuyen los precios de los productos?

```python
sns.histplot(x = "price", data = df, bins = 20, color = "lightblue");
```

```python
df.describe().T
```


¿Cuál es la relación entre el precio y la calificación promedio de los productos?

```python
sns.scatterplot(x = "price", 
                y = "mean_product_ratings_count", 
                data = df, 
                size = 10, 
                marker= "v", 
                color = "orange")


# cambiamos los nombres de los ejes
plt.xlabel("Precio")
plt.ylabel("Valoración media producto");
```

```python
sns.regplot(x = "price", 
            y = "mean_product_ratings_count", 
            data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "teal", "s": 1} # cambiamos el color y el tamaño de los puntos del scaterplot
            )


plt.xlabel("Precio")
plt.ylabel("Valoración")


plt.title("Relación entre el precio y la valoración")


plt.gca().spines['right'].set_visible(False) # quitamos la línea de la derecha
plt.gca().spines["top"].set_visible(False) # quitamos la línea de arriba;
```


¿Cómo se distribuyen los precios de los productos en función de si se utilizan anuncios promocionales (uses_ad_boosts) o no?

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='uses_ad_boosts', y='price', data=df)

plt.title('Distribución de Precios: Con vs. Sin Ad Boosts')
plt.xlabel('¿Usa Ad Boost?')
plt.ylabel('Precio')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='price', hue='uses_ad_boosts', fill=True, common_norm=False, palette='viridis')

plt.title('Densidad de Probabilidad del Precio según Uso de Anuncios')
plt.xlabel('Precio')
plt.ylabel('Densidad')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='uses_ad_boosts', y='price', data=df, split=True, inner="quart")

plt.title('Comparación de Distribución de Precios (Violin Plot)')
plt.show()
```


¿Existe una relación entre el uso de anuncios y las unidades vendidas?

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='uses_ad_boosts', y='total_units_sold', data=df)

plt.title('Unidades vendidas: Con vs. Sin Ad Boosts')
plt.xlabel('¿Usa Ad Boost?')
plt.ylabel('Unidades vendidas')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='total_units_sold', hue='uses_ad_boosts', fill=True, common_norm=False, palette='viridis')

plt.title('Unidades Vendidas según Uso de Anuncios')
plt.xlabel('Unidades Vendidas')
plt.ylabel('Densidad')
plt.show()
```

```python
# comenzamos empezando estableciendo nuestra gráfica y indicando el número de filas y columnas que queremos
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (20,5))

# creamos un boxplot, recordemos que en el caso de Seaborn para indicar en el subplot lo queremos tenemos que usar el parámetro 'ax' dentro del método de la función
# fijaos que la única diferencia con los boxplot que hemos creado previamente es que hemos añadido el parametro 'y' 
sns.boxplot(x = "uses_ad_boosts", 
            y = "total_units_sold", 
            data = df, 
            ax = axes[0], 
            palette= "mako")


# creamos nuestro primer barplot. 📌 LO QUE NOS MUESTRA ES LA MEDIA DE EDAD EN FUNCIÓN DEL ESTADO CIVIL
sns.barplot(x = "uses_ad_boosts", 
            y = "total_units_sold", 
            data = df,
            ci = None,
            ax = axes[1], 
            palette = "magma")


# añadimos título al conjunto de todas las gráficas usando '.suptitle()'
plt.suptitle("Comparación unidades vendidas y uso de publicidad")

# añadimos el 'plt.tigth_layout()' para que se ajusten los elementos de la gráfica
plt.tight_layout()
```
