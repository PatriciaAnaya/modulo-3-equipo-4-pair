4.- Promedio de unidades vendidas por tipo de envío. Nuestro objetivo es analizar cómo se correlaciona el tipo de envío (columna shipping_option_name ) con las unidades vendidas. 5.- Reemplazo de valores. A lo largo del DataFrame tenemos algunas columnas cuyos valores son cero y uno, lo que podría llevar a confusión en su interpretación. Todas estas columnas cumplen un patrón, y es que los 0 corresponden con 'No' y los 1 con 'Si'. Por esto, en este ejercicio deberéis reemplazar los 0: No y los 1: Si.

```python
diccionario_mapa = {0.: "No", 1.: "Si"} 
df_merchant["uses_ad_boosts"] = df_merchant["uses_ad_boosts"].map(diccionario_mapa)
df_merchant['uses_ad_boosts'].unique()
```

```python
df_merchant["badge_local_product"] = df_merchant["badge_local_product"].map(diccionario_mapa)
df_merchant["badge_product_quality"] = df_merchant["badge_product_quality"].map(diccionario_mapa)
df_merchant["badge_fast_shipping"] = df_merchant["badge_fast_shipping"].map(diccionario_mapa)
df_merchant["shipping_is_express"] = df_merchant["shipping_is_express"].map(diccionario_mapa)
```

6.- Calificación de la calidad del producto. Sobreescribe la columna badge_product_quality utilizando apply() para asignar "Buena" a las filas donde 'badge_product_quality' sea "No" y "Excelente" donde sea "Si".

```python
def categorizar_calidad(calidad):

    if calidad == "Si":
        return "Excelente"
    
    elif calidad == "No":
        return "Buena"
   
```

```python
# Aplicación de la función customizada con apply

df_merchant["badge_product_quality"] = df_merchant["badge_product_quality"].apply(categorizar_calidad)

df_merchant.head(5)
```

7.- Cálculo de descuento. Crea una nueva columna discount_percentage utilizando apply() que calcule el porcentaje de descuento basado en las columnas 'price' y 'retail_price', redondea los decimales a 2. Después elimina las columnas de average_discount y mean_discount

```python
# Creación de la función customizada 
def calculo_descuento(fila):

    resultado = (fila['retail_price'] - fila['price'])/fila['retail_price']*100
    return round(resultado, 2)
```

```python
# Aplicación de la función customizada con apply
df_merchant["discount_percentage"] = df_merchant.apply(calculo_descuento, axis = 1)
df_merchant.head(2)
```

```python
df_merchant.drop(["average_discount", 'mean_discount'], axis = 1, inplace = True)
df_merchant.head(2)
```

8.- Evaluación de la valoración del vendedor. Utiliza apply() para crear una nueva columna 'seller_reputation' donde los valores sean "Buena" si 'merchant_rating' es mayor o igual a 4, y "Regular" si es menor a 4. Después elimina la columna merchant_rating.

```python
def categorizar_valoracion(valoracion):

    if valoracion >=4:
        return "Buena"
    
    elif valoracion <4:
        return "Regular"
```

```python
# Aplicación de la función customizada con apply

df_merchant["seller_reputation"] = df_merchant["merchant_rating"].apply(categorizar_valoracion)

df_merchant.head(2)
```
